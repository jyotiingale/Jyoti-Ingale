```python
for feature in categorical_feat:
  data=car
  car.groupby(feature)["Year"].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('Year')
  plt.title(feature)
  plt.figure(figsize=(30,10))
  plt.show()
```

```python
for feature in numerical_feat:
  car.groupby(feature)["Year"].mean().plot()
  plt.xlabel(feature)
  plt.ylabel('Year')
  plt.title(feature)
  plt.show()
```

```python
discrete_feature=[feature for feature in numerical_feat if len(car[feature].unique())<25 and feature not in ["Id"]]
print("Discrete Variables Count: ()",format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
car[discrete_feature].head()
```

```python
for feature in discrete_feature:
  car.groupby(feature)["Year"].mean().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('Year')
  plt.title(feature)
  plt.show()
```

```python
continuous_feature=[feature for feature in numerical_feat if feature not in discrete_feature]
continuous_feature
```

```python
for feature in continuous_feature:
  car[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel("Conut")
  plt.title(feature)
  plt.show()
```
